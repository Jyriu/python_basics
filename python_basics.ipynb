Créer une petite "base de données" d'utilisateurs en utilisant une liste de dictionnaires et réaliser les actions suivantes :

Ajouter un utilisateur.
Trier les utilisateurs par âge.
Rechercher un utilisateur par nom.

```python
# Liste des utilisateurs
users = [
    {'nom': 'Sami', 'age': 27, 'email': 'bgdu94@gmail.com'},
    {'nom': 'Tarah', 'age': 27, 'email': 'bgdu95@gmail.com'}
]

# Afficher la liste des utilisateurs existants avec un format propre
print('Liste des utilisateurs :')
for user in users:
    print(f"Nom : {user['nom']}, Âge : {user['age']}, Email : {user['email']}")

# Ajouter un nouvel utilisateur
userName = input('Entrer un nom : ')
userAge = int(input('Entrer un âge : '))
userEmail = input('Entrer un email : ')
new_user = {'nom': userName, 'age': userAge, 'email': userEmail}
users.append(new_user)

# Trier les utilisateurs par âge
sortedUsers = sorted(users, key=lambda x: x['age'])

print('\nUtilisateurs triés par âge : ')
for user in sortedUsers:
    print(f"Nom : {user['nom']}, Âge : {user['age']}, Email : {user['email']}")

# Rechercher un utilisateur par son nom
userToSearch = input('\nEntrer un nom à rechercher : ')
found = False
for user in users:
    if user['nom'] == userToSearch:  # Vérifie si le nom correspond
        print(f"\nUtilisateur trouvé : Nom : {user['nom']}, Âge : {user['age']}, Email : {user['email']}")
        found = True
        break

if not found:
    print('\nUtilisateur introuvable.')
```

Créer un petit programme qui simule un système de gestion d'inventaire pour un magasin. Ce programme permettra :

D'afficher les articles disponibles.
D'ajouter ou de supprimer des articles du panier d'un utilisateur.
De calculer le total des articles dans le panier.

```python
print('Bienvenue dans notre magasin !')

# Initialisation des articles du magasin
articles = [
    {'nom': 'Chaussures', 'prix': 50},
    {'nom': 'T-shirt', 'prix': 20},
    {'nom': 'Pantalon', 'prix': 30}
]

# Affichage des articles dispo
print('\nArticles disponibles :')
for article in articles:
    print(f'{article['nom']} - {article['prix']}€')


# Choisir l'action
print('Que souhaitez-vous faire ?')

# Gestion du panier
panier = []

# Ajouter un article
def addArticle():
    chosenArticle = input('Entrez le nom de l\'article à ajouter : ').capitalize()
    for article in articles:
        if article['nom'] == chosenArticle:
            panier.append(article)
            print(f'\n{chosenArticle} a été ajouté à votre panier.')
            return
    print('Désolé, cet article n\'est pas disponible.')

# Supprimer un article
def deleteArticle():
    deletedArticle = input('Entrez le nom de l\'article à supprimer : ').capitalize()
    for article in panier:
        if article['nom'] == deletedArticle:
            panier.remove(article)
            print(f'\n{deletedArticle} a été supprimé de votre panier.')
            return
    print('\nDésolé, cet article n\'est pas dans votre panier.')

# Afficher et calculer total panier
def getPanierAndTotal():
    if len(panier) == 0:
        print('Votre panier est vide.')
    else:
        total = 0
        print('\nVotre panier contient :')
        for article in panier:
            print(f'- {article['nom']} : {article['prix']}€')
            total += article['prix']

        print(f'\nTotal à payer : {total}€')

# Menu d'action 

while True:
    print('\nArticles disponibles : ')
    for article in articles:
    print(f'{article['nom']} - {article['prix']}€')

    print('\nQue souhaitez-vous faire ?')
    print('1. Ajouter un article au panier.')
    print('2. Supprimer un article du panier.')
    print('3. Valider le panier.')
    print('4. Quitter.')

    choice = input('Entrez votre choix : ')

    if choice == '1':
        addArticle()
    elif choice == '2':
        deleteArticle()
    elif choice == '3':
        getPanierAndTotal()
    elif choice == '4':
        print('Merci pour votre visite !')
        break
    else:
        print('Choix invalide, veuillez réessayer.')
```
